```python
#HLS 脚本文件，用于批量测试HLS工程
#读取文件 ##这一段是小数部分不变，固定为10，整数部分位宽从0-10
for i in range(0,11):
    #写入.c 
    cpp_front= open("AF_front.cpp")  
    #逐行写入
    for line in cpp_front.readlines():  
        with open("AF_classifier_"+str(i+10)+"_"+str(i)+".cpp","a") as cpp:
            cpp.write(line)  
    cpp_front.close()
    
    with open("AF_classifier_"+str(i+10)+"_"+str(i)+".cpp","a") as cpp:
        cpp.write("typedef ap_fixed<"+str(i+10)+","+str(i)+"> ap_f;")
    
    cpp_end = open("AF_end.cpp")  
    #逐行写入
    for line in cpp_end.readlines():  
        with open("AF_classifier_"+str(i+10)+"_"+str(i)+".cpp","a") as cpp:
            cpp.write(line)      
    cpp_end.close()    
    #写入tcl
    with open("script.tcl","a") as tcl:
        tcl.write("\n\n\n########################\nopen_project HLS_"+str(i+10)+"_"+str(i)+"\n")
        tcl.write("set_top AF_classifier\n")
        tcl.write("add_files "+"AF_classifier_"+str(i+10)+"_"+str(i)+".cpp\n")
    
    tcl_end = open("script_end.tcl")  
    for line in tcl_end.readlines():  
        with open("script.tcl","a") as tcl:
            tcl.write(line)  
    tcl_end.close()          
```
